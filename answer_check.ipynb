```python
import numpy as np
import pandas as pd
from IPython.display import display

%matplotlib inline

data = pd.read_csv("data_without_labels.csv")
pred_zscore = pd.read_csv("data_with_prediction_zscore.csv")
pred_zscore_aspect = pd.read_csv("data_with_prediction_zscore_aspect.csv")
pred_autoencoder = pd.read_csv("data_with_prediction_autoencoder.csv")
print(sum(pred_zscore['Predicted_Labels']))
print(sum(pred_zscore_aspect['reduced_mask']))
print(sum(pred_autoencoder['prediction']))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', palette='muted', font_scale=1.5)
%matplotlib inline

conf_matrix = confusion_matrix(data['Labels'], pred_zscore['Predicted_Labels'])
plt.figure(figsize=(12, 12))
sns.heatmap(conf_matrix, xticklabels=[0, 1], yticklabels=[0, 1], annot=True, fmt="d");
plt.title("Confusion matrix")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```
