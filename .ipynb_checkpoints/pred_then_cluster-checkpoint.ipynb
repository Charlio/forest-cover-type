```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import visuals as vs

%matplotlib inline
sns.set(style='whitegrid', palette='muted', font_scale=1.5)
```

```python
data = pd.read_csv("data_without_labels.csv")
```

```python
def mad_based_outlier(points, thresh=2.5):
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff)

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return modified_z_score > thresh
```

```python
masks = mad_based_outlier(data['Elevation'], 3.2) # 3.2
print(sum(masks))
```

```python
pred_data_0 = data[masks == 0]
pred_data_1 = data[masks == 1]
print(pred_data_0.shape)
print(pred_data_1.shape)
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(data['Labels'], masks)
plt.figure(figsize=(12, 12))
sns.heatmap(conf_matrix, xticklabels=[0, 1], yticklabels=[0, 1], annot=True, fmt="d");
plt.title("Confusion matrix")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```

```python
from random import sample

indices = sample(list(range(pred_data_0.shape[0])), pred_data_1.shape[0])

print(len(indices))
pred_data_0_samples = pred_data_0.iloc[indices]
print(pred_data_0_samples.shape)
```

```python
balanced_data = pd.concat([pred_data_1, pred_data_0_samples])
print(balanced_data.shape)
```

```python
balanced_labels = balanced_data['Labels']
del balanced_data['Labels']
del balanced_data['Observation_ID']
```

```python
del data['Labels']
del data['Observation_ID']
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(balanced_data);
```

```python
pca = PCA(n_components=3)
pca.fit(balanced_data)

reduced_data = pca.transform(balanced_data)

reduced_data = pd.DataFrame(reduced_data, columns = ['Dimension 1', 'Dimension 2', 'Dimension 3'])
```

```python
pca_results = vs.pca_results(balanced_data, pca)
```

```python
vs.biplot(balanced_data, reduced_data, pca);
```

```python
minor_data = balanced_data[balanced_labels == 1]
pca_samples = pca.transform(minor_data)
```

```python
from sklearn.mixture import GaussianMixture

clusterer = GaussianMixture(n_components=2).fit(reduced_data)

preds = clusterer.predict(reduced_data)

centers = clusterer.means_

sample_preds = clusterer.predict(pca_samples)

from sklearn.metrics import silhouette_score

score = silhouette_score(reduced_data, preds)
```

```python
print(centers)
print(score)
```

```python
vs.cluster_results(reduced_data, preds, centers, pca_samples)
```
