## z-score on Elevation

```python
def mad_based_outlier(points, thresh=2.5):
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff)

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return modified_z_score > thresh
```

```python
close_masks = mad_based_outlier(data['Elevation'], 3.2)
print(sum(close_masks))
data['Predicted_Labels'] = close_masks & (data['Elevation'] < 3000)
print(sum(data['Predicted_Labels']))
inital_predictions = data[['Observation_ID', 'Predicted_Labels']]
inital_predictions.to_csv('data_with_prediction_zscore.csv')
```

```python
reduced_data = data[masks == 1]
reduced_data = reduced_data[reduced_data['Elevation'] < 3000]
print(reduced_data.shape)
```

```python
label_masks = (masks == 1) & (data['Elevation'] < 3000)
data['close_mask'] = close_masks
reduced_labels = data[label_masks == 1]['close_mask']
print(reduced_labels.shape)
print(sum(reduced_labels))
```

### Clustering

```python
reduced_features = [
    'Elevation',
    'Aspect',
    #'Slope'
]
```

```python
aspect_data = reduced_data[reduced_features]
#normalized_data=(reduced_data-reduced_data.mean())/reduced_data.std()
```

```python
pca = PCA()
pca.fit(aspect_data)
pca_results = vs.pca_results(aspect_data, pca)
```

```python
pca = PCA(n_components=2)
pca.fit(aspect_data)
pca_data = pca.transform(aspect_data)
pca_data = pd.DataFrame(pca_data, columns = ['Dimension 1', 'Dimension 2'])
pca_samples = pca.transform(aspect_data[reduced_labels == 1])
```

```python
vs.biplot(aspect_data, pca_data, pca);
```

```python
from sklearn.mixture import GaussianMixture
clusterer = GaussianMixture(n_components=2).fit(pca_data)
```

```python
preds = clusterer.predict(pca_data)
centers = clusterer.means_
print(centers)
```

```python
vs.cluster_results(pca_data, preds, centers, pca_samples)
```
